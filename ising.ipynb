```python
import numpy as np
from math import exp
import random

# parameters
size = 10
T = 2.5

# spin array
s = np.zeros((size, size), dtype = int)
```

```python
def initialize():
    for i in range(size):
        for j in range(size):
            # random spin +/- 1
            if random.random() < 0.5:
                s[i, j] = 1
            else:
                s[i, j] = -1
            colorsquare(i, j)
```

```python
def deltaU(i, j):
    '''
    compute energy difference for flipping spin at (i, j) using periodic boundary conditions
    '''
    top = s[(i-1) % size, j]
    bottom = s[(i+1) % size, j]
    left = s[i, (j-1) % size]
    right = s[i, (j+1) % size]

    Ediff = 2 * s[i, j] * (top + bottom + left + right)
    return Ediff
```

```python
def colorsquare(i, j):
    # simple console visualization
    # +1 → "⬜",  -1 → "⬛"
    char = "⬜" if s[i, j] == 1 else "⬛"

    # print the whole grid after each update
    # (comment this out if too slow)
    for row in s:
        print("".join("⬜" if x == 1 else "⬛" for x in row))
    print()
```

```python
def main():
    initialize()

    steps = 100 * size * size
    for _ in range(steps):
        # choose random site
        i = random.randrange(size)
        j = random.randrange(size)

        Ediff = deltaU(i, j)


        if Ediff <= 0:
            s[i, j] *= -1
            colorsquare(i, j)
        else:
            if random.random() < exp(-Ediff / T):
                s[i, j] *= -1
                colorsquare(i, j)

if __name__ == "__main__":
    main()
```
